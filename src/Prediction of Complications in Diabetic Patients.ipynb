### 生成模拟数据：糖尿病确诊日期、冠心病确诊日期

```python
from datetime import datetime, timedelta

num_samples = len(df_filled)  # 获取数据集样本数量

# 生成糖尿病日期随机数据
df['糖尿病确诊日期'] = pd.to_datetime({
    'year': np.random.randint(2000, 2023, size=num_samples),
    'month': np.random.randint(1, 13, size=num_samples),
    'day': np.random.randint(1, 29, size=num_samples)
})

# 定义函数生成随机日期
def generate_random_date(start_date, end_date):
    start_datetime = datetime.strptime(start_date, '%Y-%m-%d')
    end_datetime = datetime.strptime(end_date, '%Y-%m-%d')
    random_days = np.random.randint((end_datetime - start_datetime).days + 1)
    return (start_datetime + timedelta(days=random_days)).strftime('%Y-%m-%d')

# 添加冠心病日期列
df['冠心病确诊日期'] = np.where(df['冠心病'] == 1, [generate_random_date('2015-01-01', '2025-12-31') for _ in range(len(df))], np.nan)
```

```python
#接下来，利用贝叶斯模型平均方法进行预测（选用的基础模型为逻辑回归、决策树和随机森林）
#步骤：
#1. 训练每个模型并获取其预测概率；
#2. 使用预测概率来估计每个模型的后验概率；
#3. 根据后验概率计算每个模型的权重。
# 训练模型并获取预测概率

logistic_probs = logistic_model.predict_proba(X_test)[:, 1]

decision_tree_probs = decision_tree_model.predict_proba(X_test)[:, 1]

random_forest_probs = random_forest_model.predict_proba(X_test)[:, 1]

# 估计每个模型的后验概率
logistic_posterior = np.mean(logistic_probs)
decision_tree_posterior = np.mean(decision_tree_probs)
random_forest_posterior = np.mean(random_forest_probs)

# 计算每个模型的权重
total_posterior = logistic_posterior + decision_tree_posterior + random_forest_posterior
logistic_weight = logistic_posterior / total_posterior
decision_tree_weight = decision_tree_posterior / total_posterior
random_forest_weight = random_forest_posterior / total_posterior

print("逻辑回归模型权重：", logistic_weight)
print("决策树模型权重：", decision_tree_weight)
print("随机森林模型权重：", random_forest_weight)
```
